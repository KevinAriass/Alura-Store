# 4. Productos más y menos vendidos

```python
producto_mas_vendido_tienda_1 = productos_vendidos_tienda_1.idxmax()
producto_menos_vendido_tienda_1 = productos_vendidos_tienda_1.idxmin()
producto_mas_vendido_tienda_2 = productos_vendidos_tienda_2.idxmax()
producto_menos_vendido_tienda_2 = productos_vendidos_tienda_2.idxmin()
producto_mas_vendido_tienda_3 = productos_vendidos_tienda_3.idxmax()
producto_menos_vendido_tienda_3 = productos_vendidos_tienda_3.idxmin()
producto_mas_vendido_tienda_4 = productos_vendidos_tienda_4.idxmax()
producto_menos_vendido_tienda_4 = productos_vendidos_tienda_4.idxmin()
```

```python
print(f'A continuación encontrará el producto más y menos vendido por tienda.')

nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

productos_mas_vendidos = [
    productos_vendidos_tienda_1.idxmax(),
    productos_vendidos_tienda_2.idxmax(),
    productos_vendidos_tienda_3.idxmax(),
    productos_vendidos_tienda_4.idxmax()
]

cantidades_mas_vendidos = [
    productos_vendidos_tienda_1.max(),
    productos_vendidos_tienda_2.max(),
    productos_vendidos_tienda_3.max(),
    productos_vendidos_tienda_4.max()
]

productos_menos_vendidos = [
    productos_vendidos_tienda_1.idxmin(),
    productos_vendidos_tienda_2.idxmin(),
    productos_vendidos_tienda_3.idxmin(),
    productos_vendidos_tienda_4.idxmin()
]

cantidades_menos_vendidos = [
    productos_vendidos_tienda_1.min(),
    productos_vendidos_tienda_2.min(),
    productos_vendidos_tienda_3.min(),
    productos_vendidos_tienda_4.min()
]

x = np.arange(len(nombres_tiendas))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, cantidades_mas_vendidos, width, label='Más Vendido')
rects2 = ax.bar(x + width/2, cantidades_menos_vendidos, width, label='Menos Vendido')

# Añadir etiquetas de texto
ax.set_ylabel('Cantidad de unidades vendidas')
ax.set_title('Cantidad de productos más y menos vendidos por Tienda')
ax.set_xticks(x)
ax.set_xticklabels(nombres_tiendas)
ax.legend()

def autolabel(rects, productos):
    """Attach a text label above each bar in *rects*, displaying its height and product name."""
    for i, rect in enumerate(rects):
        height = rect.get_height()
        ax.annotate(f'{height} ({productos[i]})',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1, productos_mas_vendidos)
autolabel(rects2, productos_menos_vendidos)

fig.tight_layout()

plt.show()

# Tienda 1
producto_mas_vendido_tienda_1 = productos_vendidos_tienda_1.idxmax()
cantidad_mas_vendido_tienda_1 = productos_vendidos_tienda_1[producto_mas_vendido_tienda_1]
producto_menos_vendido_tienda_1 = productos_vendidos_tienda_1.idxmin()
cantidad_menos_vendido_tienda_1 = productos_vendidos_tienda_1[producto_menos_vendido_tienda_1]

print(f'Tienda 1:')
print(f'  Los {producto_mas_vendido_tienda_1} son el producto más vendido de la tienda, con una cantidad de: {cantidad_mas_vendido_tienda_1} unidades vendidas')
print(f'  Por otro lado los {producto_menos_vendido_tienda_1} son el menos vendido, con una cantidad de: {cantidad_menos_vendido_tienda_1} unidades vendidas')

# Tienda 2
producto_mas_vendido_tienda_2 = productos_vendidos_tienda_2.idxmax()
cantidad_mas_vendido_tienda_2 = productos_vendidos_tienda_2[producto_mas_vendido_tienda_2]
producto_menos_vendido_tienda_2 = productos_vendidos_tienda_2.idxmin()
cantidad_menos_vendido_tienda_2 = productos_vendidos_tienda_2[producto_menos_vendido_tienda_2]

print(f'Tienda 2:')
print(f'  En esta tienda también {producto_mas_vendido_tienda_2} son el producto más vendido, con una cantidad de: {cantidad_mas_vendido_tienda_2} unidades vendidas')
print(f'  Al igual que en la tienda anterior los {producto_menos_vendido_tienda_2} son los de menos venta, con una cantidad de: {cantidad_menos_vendido_tienda_2} unidades vendidas')

# Tienda 3
producto_mas_vendido_tienda_3 = productos_vendidos_tienda_3.idxmax()
cantidad_mas_vendido_tienda_3 = productos_vendidos_tienda_3[producto_mas_vendido_tienda_3]
producto_menos_vendido_tienda_3 = productos_vendidos_tienda_3.idxmin()
cantidad_menos_vendido_tienda_3 = productos_vendidos_tienda_3[producto_menos_vendido_tienda_3]

print(f'Tienda 3:')
print(f'  Los {producto_mas_vendido_tienda_3} lideran las ventas de esta tienda, con una cantidad de: {cantidad_mas_vendido_tienda_3} unidades vendidas')
print(f'  En este caso los {producto_menos_vendido_tienda_3} son los que muestran ventas más bajas, con una cantidad de: {cantidad_menos_vendido_tienda_3} unidades vendidas')

# Tienda 4
producto_mas_vendido_tienda_4 = productos_vendidos_tienda_4.idxmax()
cantidad_mas_vendido_tienda_4 = productos_vendidos_tienda_4[producto_mas_vendido_tienda_4]
producto_menos_vendido_tienda_4 = productos_vendidos_tienda_4.idxmin()
cantidad_menos_vendido_tienda_4 = productos_vendidos_tienda_4[producto_menos_vendido_tienda_4]

print(f'Tienda 4:')
print(f'  Al igual que en las otras tiendas los {producto_mas_vendido_tienda_4} son el producto más vendido, con una cantidad de: {cantidad_mas_vendido_tienda_4} unidades vendidas')
print(f'  Los {producto_menos_vendido_tienda_4} muestran pocas ventas en esta tienda también, con una cantidad de: {cantidad_menos_vendido_tienda_4} unidades vendidas')
```
